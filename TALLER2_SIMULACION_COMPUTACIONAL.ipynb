```python
def linear_congruential_generator(a, c, m, X0, n):
    values = []
    Xn = X0
    for _ in range(n):
        X_next = (a * Xn + c) % m
        values.append(X_next / float(m))
        Xn = X_next
    return values

# Parámetros
a, c, m, X0 = 8,13,97, 8
n = 500  # Número de valores a generar
lcg_values = linear_congruential_generator(a, c, m, X0, n)
print(lcg_values)
```

```python
def min_std_generator(a, m, X0, n):
    values = []
    Xn = X0
    q = m // a
    r = m % a
    for _ in range(n):
        Xn = ((a * (Xn % q)) - (r * (Xn // q)))
        if Xn < 0:
            Xn = Xn + m
        values.append(Xn / m)
    return values

a, m, X0 = 7**5, 2**31 - 1, 32423
n = 500
min_std_values = min_std_generator(a, m, X0, n)
print(min_std_values)
```

```python
def chi_square_test(data, num_intervals):
    observed_frequencies, _ = np.histogram(data, bins=num_intervals)
    expected_frequencies = np.ones_like(observed_frequencies) * (len(data) / num_intervals)
    chi_square_stat = chisquare(observed_frequencies, f_exp=expected_frequencies).statistic
    return chi_square_stat

num_intervals = 10
alpha = 0.05  # Nivel de significancia

chi_stat_lcg = chi_square_test(lcg_values, num_intervals)
chi_stat_minstd = chi_square_test(min_std_values, num_intervals)
df = num_intervals - 1  # Grados de libertad
chi_critical_value = chi2.ppf(1 - alpha, df)
def interpretar_chi_cuadrado(chi_stat, chi_critical_value):
    return "Aceptar H0: Distribución uniforme" if chi_stat <= chi_critical_value else "Rechazar H0: No es una distribución uniforme"

print("Resultados de la prueba Chi-cuadrado:")
print(f"{'Generador':<30}{'Chi-cuadrado Estadístico':<30}{'Valor Crítico de Chi-cuadrado':<30}{'Interpretación':<30}")
print(f"{'LCG':<30}{chi_stat_lcg:<30.2f}{chi_critical_value:<30.2f}{interpretar_chi_cuadrado(chi_stat_lcg, chi_critical_value):<30}")
print(f"{'Min Std':<30}{chi_stat_minstd:<30.2f}{chi_critical_value:<30.2f}{interpretar_chi_cuadrado(chi_stat_minstd, chi_critical_value):<30}")

```

```python
def kolmogorov_smirnov_test(data, alpha=0.05):
    n = len(data)
    statistic, p_value = kstest(data, 'uniform')
    K_alpha = 1.36  # Valor para un nivel de significancia de 0.05
    critical_value = K_alpha / np.sqrt(n)
    return statistic, critical_value


ks_stat_lcg, ks_critical_lcg = kolmogorov_smirnov_test(lcg_values)
ks_stat_minstd, ks_critical_minstd = kolmogorov_smirnov_test(min_std_values)

def interpretar_ks(statistic, critical_value):
    return "Aceptar H0: Distribución uniforme" if statistic <= critical_value else "Rechazar H0: No es una distribución uniforme"

print("Resultados de la prueba Kolmogorov-Smirnov:")
print(f"{'Generador':<30}{'K-S Estadístico':<30}{'Valor Crítico de K-S':<30}{'Interpretación':<30}")
print(f"{'LCG':<30}{ks_stat_lcg:<30.4f}{ks_critical_lcg:<30.4f}{interpretar_ks(ks_stat_lcg, ks_critical_lcg):<30}")
print(f"{'Min Std':<30}{ks_stat_minstd:<30.4f}{ks_critical_minstd:<30.4f}{interpretar_ks(ks_stat_minstd, ks_critical_minstd):<30}")
```

```python
import numpy as np

def runs_test(values, alpha=0.05):
    runs = 0
    n1 = 0
    n2 = 0
    runs_serie = ""

    # Calcular + y -
    for i in range(1, len(values)):
        if values[i] > values[i-1]:
            n1 += 1
            runs_serie += "+"
        else:
            n2 += 1
            runs_serie += "-"

    # Calcular corridas (+ o -)
    prev_run = runs_serie[0]
    for i in range(1, len(runs_serie)):
        if runs_serie[i] != prev_run:
            runs += 1
            prev_run = runs_serie[i]
    runs += 1  # La última corrida

    # Calcular estadísticas
    mu = (2*n1*n2) / (n1 + n2) + 1
    sigma_sq = (2*n1*n2*(2*n1*n2 - n1 - n2)) / ((n1 + n2)**2 * (n1 + n2 - 1))
    z = (runs - mu) / np.sqrt(sigma_sq)

    # Valor crítico para la prueba bilateral
    z_critical = norm.ppf(1 - alpha/2)

    print(f"Secuencia de + y -: {runs_serie}")
    print(f"Número total de corridas: {runs}")

    return {
        'n1': n1,
        'n2': n2,
        'runs': runs,
        'mu': mu,
        'sigma_sq': sigma_sq,
        'z': z,
        'z_critical': z_critical,
        'result': abs(z) < z_critical
    }


result_min_std_values = runs_test(min_std_values)
results_lcg = runs_test(lcg_values)

print("Prueba de corridas para LCG:", results_lcg)

print("Prueba de corridas para STD - MIN:", result_min_std_values)
```

```python
import numpy as np
from scipy.stats import chisquare, chi2


# Función para el test de series
def series_test(data, k):
    n = len(data) - 1
    fe = (n / (k ** 2))

    observed_matrix = np.zeros((k, k))
    intervals = np.linspace(0, 1, k + 1)

    for i in range(n):
        for j in range(k):
            if intervals[j] <= data[i] < intervals[j+1]:
                for m in range(k):
                    if intervals[m] <= data[i+1] < intervals[m+1]:
                        observed_matrix[j][m] += 1

    observed = observed_matrix.flatten()
    expected = np.full(k**2, fe)
    if not np.isclose(observed.sum(), expected.sum(), rtol=1e-5):
        print("La suma de las frecuencias observadas no coincide con la esperada.")
        difference = expected.sum() - observed.sum()
        expected += difference / (k**2)

    chi_stat, p_value = chisquare(observed, f_exp=expected)

    return observed_matrix, chi_stat, p_value


from scipy.stats import chisquare, chi2

k = 5  # Número de intervalos
# Realiza la prueba de series
observed_matrix_lcg, chi_stat_lcg, p_value_lcg = series_test(lcg_values, k)
observed_matrix_minstd, chi_stat_minstd, p_value_minstd = series_test(min_std_values, k)

# Grados de libertad para la prueba de chi-cuadrado
degrees_of_freedom = k**2 - 1

# Valor crítico de chi-cuadrado al nivel de significancia deseado
alpha = 0.05
chi_crit = chi2.ppf(1 - alpha, degrees_of_freedom)

# Imprime los resultados
print("LCG - Matriz de Observados:")
print(observed_matrix_lcg)
print("LCG - Estadístico de Chi-Cuadrado:", chi_stat_lcg)
print("LCG - Valor crítico de Chi-Cuadrado:", chi_crit)

print("\nMinStd - Matriz de Observados:")
print(observed_matrix_minstd)
print("MinStd - Estadístico de Chi-Cuadrado:", chi_stat_minstd)
print("MinStd - Valor crítico de Chi-Cuadrado:", chi_crit)

# Decisión basada en el valor crítico
if chi_stat_lcg > chi_crit:
    print("\nLCG: Rechazar la hipótesizar de independecia.")
else:
    print("\nLCG: Acepta la hipotesis de independencia.")

if chi_stat_minstd > chi_crit:
    print("MinStd: Rechazar la hipótes de independecia")
else:
    print("MinStd: Acepta la hipotesis de independencia.")
```

```python
from scipy.stats import chi2

def categorize_in_threes(values):
    patterns = {'todos_diferentes': 0, 'un_par': 0, 'tres_iguales': 0}
    for i in range(0, len(values), 3):
        hand = sorted(values[i:i+3])
        if len(hand) < 3: continue
        counts = list(map(hand.count, hand))
        pattern = (len(set(counts)), max(counts))
        if pattern == (3, 1):
            patterns['todos_diferentes'] += 1
        elif pattern == (1, 3):
            patterns['tres_iguales'] += 1
        elif pattern == (2, 2):
            patterns['un_par'] += 1
    return patterns

# Función para realizar la prueba chi-cuadrado
def perform_chi_squared_test(values, n):
    patterns = categorize_in_threes(values)
    observed = list(patterns.values())
    probabilities = {'todos_diferentes': 0.720, 'un_par': 0.270, 'tres_iguales': 0.010}
    expected = [n * probabilities[pattern] for pattern in patterns]
    chi_squared = sum((o - e) ** 2 / e for o, e in zip(observed, expected))
    df = len(patterns) - 1
    critical_value = chi2.ppf(1 - alpha, df)
    return chi_squared, critical_value

n_groups = n // 3

# Prueba para LCG
chi_squared_lcg, critical_value_lcg = perform_chi_squared_test(lcg_values, n_groups)
print("LCG Chi cuadrado:", chi_squared_lcg)
print("LCG Valor crítico:", critical_value_lcg)
print("Hipótesis nula LCG", "rechazada" if chi_squared_lcg > critical_value_lcg else "aceptada")

# Prueba para MinStd
chi_squared_minstd, critical_value_minstd = perform_chi_squared_test(min_std_values, n_groups)
print("MinStd Chi cuadrado:", chi_squared_minstd)
print("MinStd Valor crítico:", critical_value_minstd)
print("Hipótesis nula MinStd", "rechazada" if chi_squared_minstd > critical_value_minstd else "aceptada")

```

```python
from scipy.stats import chi2

# Probabilidades esperadas para cada patrón
probabilities = {
    'todos_diferentes': 0.3024,
    'un_par': 0.5040,
    'dos_pares': 0.1080,
    'tercia': 0.0720,
    'full': 0.0090,
    'poker': 0.0045,
    'quintilla': 0.0001
}

# Función para calcular la prueba de chi-cuadrado
def calculate_chi_squared(observed_counts, expected_probabilities, total_samples):
    expected_counts = [total_samples * p for p in expected_probabilities]
    chi_squared_stat = sum((o - e) ** 2 / e for o, e in zip(observed_counts, expected_counts))
    return chi_squared_stat

# Función para realizar la prueba de chi cuadrado para una secuencia de números
def poker_test(values):
    patterns = {'todos_diferentes': 0, 'un_par': 0, 'dos_pares': 0, 'tercia': 0,
                'full': 0, 'poker': 0, 'quintilla': 0}
    for i in range(0, len(values), 5):
        hand = sorted(values[i:i+5])
        if len(hand) < 5:
            continue
        counts = list(map(hand.count, hand))
        pattern = (len(set(counts)), max(counts))
    chi_squared_stat = calculate_chi_squared(list(patterns.values()), list(probabilities.values()), len(values) // 5)
    alpha = 0.05
    degrees_of_freedom = len(patterns) - 1
    critical_value = chi2.ppf(1 - alpha, degrees_of_freedom)
    if chi_squared_stat > critical_value:
        decision = 'rechazada'
    else:
        decision = 'aceptada'
    return chi_squared_stat, critical_value, decision

# Realiza la prueba para LCG
chi_squared_lcg, critical_value_lcg, decision_lcg = poker_test(lcg_values)
print("LCG Chi cuadrado:", chi_squared_lcg)
print("LCG Valor crítico:", critical_value_lcg)
print(f"Hipótesis nula LCG {decision_lcg}")

# Realiza la prueba para MinStd
chi_squared_minstd, critical_value_minstd, decision_minstd = poker_test(min_std_values)
print("MinStd Chi cuadrado:", chi_squared_minstd)
print("MinStd Valor crítico:", critical_value_minstd)
print(f"Hipótesis nula MinStd {decision_minstd}")
```
